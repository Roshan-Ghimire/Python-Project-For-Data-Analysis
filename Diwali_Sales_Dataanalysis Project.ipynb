# Exploratory Data Analysis

##### Analysis Based on Gender

```python
data=df['Gender'].value_counts().reset_index()
data
```

```python
plt.pie(data['count'],labels=['Female','Male'],data=data,autopct='%1.1f%%',startangle=45,explode=(0.1,0),shadow=False)
```

```python
## Total Amount Spend by Males and Female
tot=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
tot
```

```python
plt.pie('Amount',labels='Gender',data=tot,autopct='%1.2f%%',startangle=50,shadow=False,radius=1.6,explode=(0.2,0),pctdistance=0.8)
circle=plt.Circle((0,0),1,color='white')
c=plt.gcf()
c.gca().add_artist(circle)
```

```python
orders=df.groupby(['Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
```

```python
orders
```

```python
bp=sp.barplot(x='Gender',y='Orders',data=orders)
for bar in bp.containers:
    bp.bar_label(bar)
plt.title("Total Orders:Female Vs Male")    
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.
And We can also see that Most of the orders in our sales in given by Female as Comparison to Male

#### Analysis Based on Ages

```python
# Total number of Customer in each Age Group
sp.set(rc={'figure.figsize':(12,5)})
Age=sp.countplot(x='Age Group',data=df)
for bar in Age.containers:
    Age.bar_label(bar)
plt.title("Count Of Age Groups") 
```

```python
# Fing The total number of male and Female as per each age group
cate=df.groupby(['Age Group', 'Gender'], as_index=False)['Gender'].value_counts()
```

```python
cate
```

```python
sp.set(rc={'figure.figsize':(12,5)})
cos=sp.countplot(x='Age Group',data=df,hue='Gender')
for bar in cos.containers:
    cos.bar_label(bar)
plt.title('Age Group: Male Vs Female')    
```

```python
# Fing the total amount spent by customer of each age group
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
Age_Sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sp.set(rc={'figure.figsize':(16,5)})
Age_Sales_Bar=sp.barplot(x='Age Group',y='Amount',data=Age_Sales)
for bar in Age_Sales_Bar.containers:
    Age_Sales_Bar.bar_label(bar,fmt='%.0f')
plt.title("Age Group Vs Amount")    
```

```python
# Find Which Gender has spend most money in each Age Group
age_amount_gender=df.groupby(['Age Group','Gender'])['Amount'].sum().reset_index()
age_amount_gender
```

```python
sp.set(rc={'figure.figsize':(22,5)})
bardata=sp.barplot(x='Age Group',y='Amount',data=age_amount_gender,hue='Gender')
for bar in bardata.containers:
    bardata.bar_label(bar,fmt='%.0f')
```

From The above graph,we can say that Female From Each Age Group has spent the most money

## Analysis Of State


```python
#Find the Top 10 state with most Orders
df.columns
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

# Product_Category

```python
## Find the total order made for each category of product
df.groupby(['Product_Category'],as_index=False)['Product_Category'].value_counts().sort_values(by='count',ascending=False)
```

```python
sp.set(rc={'figure.figsize':(24,5)})
category_count=sp.countplot(x='Product_Category',data=df)
for bar in category_count.containers:
    category_count.bar_label(bar)
    
```

```python
## find the top 10 selling Products_Category
Product_amount=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
Product_amount
plt.figure(figsize=(20, 4))
sp.barplot(x='Product_Category',y='Amount',data=Product_amount)
```

```python
Fivemostselling_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
plt.figure(figsize=(8, 8))
plt.pie('Amount', labels='Product_Category',data=Fivemostselling_product,autopct='%1.1f%%', radius=1,startangle=110,explode=(0.2,0.1,0.1,0.1,0.1),pctdistance=0.75)
circle = plt.Circle((0, 0), 0.6, color='white', fc='white')
plt.gca().add_artist(circle)
plt.title('Top Five Most Selling Product')

```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
